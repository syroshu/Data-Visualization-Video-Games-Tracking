```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('games.csv')
df
```

```python
region = df.drop(columns = ['Rank', 'Year', 'Genre', 'Name', 'Platform', 'Publisher'])
region
```

```python
publisher = df.drop(columns = ['Rank', 'Year', 'Genre', 'Name'])
publisher
```

```python
publisher = publisher.groupby('Publisher').sum()
publisher
```

```python
publisher = publisher.sort_values(by=['Global_Sales'], ascending=False).head()
publisher
```

```python
publisher = publisher.rename(index={'Electronic Arts':'EA', 'Sony Computer Entertainment': 'Sony'})
publisher
```

```python
publisher = publisher.sort_values(by=['Global_Sales'], ascending=True)
publisher
```

```python
plt.style.use("bmh")
```

```python
fig, ax = plt.subplots()
ax.bar(publisher.index, publisher['Global_Sales'], color=(0.1,0.6,0.5))
ax.set_xlabel('Publisher',labelpad=10, color=(0.1,0.6,0.5))
ax.set_ylabel("Global Sales Rate", labelpad=10, color=(0.1,0.6,0.5))
plt.show()
```
